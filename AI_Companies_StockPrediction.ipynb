```python
#Importing necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from datetime import datetime
```

```python
# ----------------------------------------------
#  Fetch Historical Stock Price Data
# ----------------------------------------------

# Importing libraries again (if needed)
import yfinance as yf
import pandas as pd

# Define the list of top 15 AI-investing public companies with their ticker symbols
tickers = [
    'MSFT',  # Microsoft
    'GOOGL', # Alphabet
    'NVDA',  # Nvidia
    'META',  # Meta
    'AMZN',  # Amazon
    'AAPL',  # Apple
    'TSLA',  # Tesla
    'IBM',   # IBM
    'AMD',   # AMD
    'CRM',   # Salesforce
    'ORCL',  # Oracle
    'INTC',  # Intel
    'PLTR',  # Palantir
    'BABA',  # Alibaba
    'TSM'    # Taiwan Semiconductor
]

# Set the start and end dates for data collection
start_date = '2021-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')  # Today's date

# Create an empty dictionary to store data for each ticker
stock_data = {}

# Loop through each ticker to download data using yfinance
for ticker in tickers:
    print(f"Downloading data for {ticker}...")
    data = yf.download(ticker, start=start_date, end=end_date)
    stock_data[ticker] = data
    # Optional: Save each to CSV for backup and offline EDA if needed
    data.to_csv(f"{ticker}_stock_data.csv")

print("✅ Data download completed for all companies.")
```

📈 What is CAGR?

CAGR (Compound Annual Growth Rate):

Measures average annual growth rate over a period, accounting for compounding.
Helps you understand how much an investment grew each year on average over the period (2021–present).
✅ A higher CAGR means strong consistent growth.

⚖️ What is Volatility?

Measures how much a stock's price fluctuates around its mean.
We will use standard deviation of daily returns as a volatility measure.
Higher volatility = higher risk, but also higher potential reward.

```python
# STEP 3 – PART 1 (Final): Closing Price & 30-Day MA with high-contrast styling

import matplotlib.pyplot as plt
import pandas as pd

for ticker in tickers:
    data = stock_data[ticker].copy()
    data.index = pd.to_datetime(data.index)           # ensure datetime index

    price = data['Close']
    ma30  = price.rolling(window=30).mean()            # 30-day MA

    plt.figure(figsize=(12, 6))
    # raw price
    plt.plot(
        price.index, price,
        label='Close Price',
        color='tab:blue',
        linewidth=1,
    )
    # 30-day MA
    plt.plot(
        ma30.index, ma30,
        label='30-day MA',
        color='tab:orange',
        linewidth=3,
        linestyle='--',
    )

    plt.title(f"{ticker} Closing Price & 30-Day Moving Average", fontsize=16)
    plt.xlabel("Date", fontsize=14)
    plt.ylabel("Price (USD)", fontsize=14)
    plt.legend(loc='upper left', fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()
```

```python
# ----------------------------------------------
# STEP 3 – PART 2: Calculate CAGR & Volatility and Build a Sorted Summary Table
# ----------------------------------------------

import numpy as np
import pandas as pd

# 1) Compute CAGR and annualized volatility as pure floats
cagr_list = []
vol_list  = []

for ticker in tickers:
    df    = stock_data[ticker]
    start = df['Close'].iloc[0]            # first closing price
    end   = df['Close'].iloc[-1]           # latest closing price

    # a) CAGR
    days = (df.index[-1] - df.index[0]).days
    years = days / 365.25
    cagr = (end / start) ** (1/years) - 1
    cagr_list.append(float(cagr))          # ensure Python float

    # b) Annualized volatility
    returns = df['Close'].pct_change().dropna()
    vol = returns.std() * np.sqrt(252)
    vol_list.append(float(vol))            # ensure Python float

# 2) Build the raw summary DataFrame (unsorted)
summary_df = pd.DataFrame({
    'Ticker':         tickers,
    'CAGR (%)':       [round(x * 100, 2) for x in cagr_list],
    'Volatility (%)': [round(x * 100, 2) for x in vol_list]
})

# 3) Sort by CAGR descending using NumPy argsort (no pandas Series compare!)
order = np.argsort(-summary_df['CAGR (%)'].values)
summary_df = summary_df.iloc[order].reset_index(drop=True)

# 4) Display final, ranked table
print("✅ AI-Investing Stocks: CAGR vs. Volatility (ranked by CAGR)")
display(summary_df)
```

```python
# ----------------------------------------------
# Visualizing your ranked CAGR vs Volatility table
# ----------------------------------------------

import matplotlib.pyplot as plt

# Convert your columns to decimal for accurate axis scaling
cagr_decimal = summary_df['CAGR (%)'] / 100
vol_decimal = summary_df['Volatility (%)'] / 100
tickers = summary_df['Ticker']

# Set up the plot
plt.figure(figsize=(10, 7))

# Scatter plot
plt.scatter(cagr_decimal, vol_decimal, s=150, alpha=0.8, edgecolor='black')

# Annotate each point
for i, ticker in enumerate(tickers):
    plt.annotate(
        ticker,
        (cagr_decimal.iloc[i], vol_decimal.iloc[i]),
        textcoords="offset points",
        xytext=(5, 5),
        ha='left',
        fontsize=9
    )

# Titles and labels
plt.title('CAGR vs Volatility of AI-Investing Public Companies', fontsize=16)
plt.xlabel('CAGR (as decimal)', fontsize=14)
plt.ylabel('Volatility (as decimal)', fontsize=14)

# Grid for readability
plt.grid(True, linestyle='--', alpha=0.5)

# Add quadrant average reference lines
plt.axhline(y=vol_decimal.mean(), color='grey', linestyle='--', linewidth=1, label=f'Avg Volatility: {vol_decimal.mean():.2f}')
plt.axvline(x=cagr_decimal.mean(), color='grey', linestyle='--', linewidth=1, label=f'Avg CAGR: {cagr_decimal.mean():.2f}')

# Optional: add legend for quadrant lines
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------
# FINAL CLEAN FIX: Clean Sharpe Table with Pure Floats
# ----------------------------------------------

# Use risk-free rate
risk_free_rate = 0.04

# Lists for clean floats
annual_return_list = []
volatility_list = []

# Compute annual return and volatility cleanly
for ticker in tickers:
    df = stock_data[ticker]
    daily_returns = df['Close'].pct_change().dropna()

    annual_return = daily_returns.mean() * 252
    volatility = daily_returns.std() * np.sqrt(252)

    annual_return_list.append(round(float(annual_return * 100), 2))  # clean float
    volatility_list.append(round(float(volatility * 100), 2))        # clean float

# Ensure CAGR as clean list
cagr_list_clean = summary_df['CAGR (%)'].astype(float).round(2).tolist()

# Calculate Sharpe Ratio as clean flat list
returns_decimal = np.array(annual_return_list) / 100
volatility_decimal = np.array(volatility_list) / 100
sharpe_ratio = ((returns_decimal - risk_free_rate) / volatility_decimal).round(2)
sharpe_ratio_list = sharpe_ratio.tolist()  # ensure list of floats

# Build the final clean DataFrame
sharpe_df = pd.DataFrame({
    'Ticker': tickers,
    'CAGR (%)': cagr_list_clean,
    'Annual Return (%)': annual_return_list,
    'Volatility (%)': volatility_list,
    'Sharpe Ratio': sharpe_ratio_list
})

# Sort and display
sharpe_df_sorted = sharpe_df.sort_values(by='Sharpe Ratio', ascending=False).reset_index(drop=True)

print("✅ ✅ FINAL CLEAN Sharpe Ratio Analysis Table – Fully Corrected")
display(sharpe_df_sorted)
```

```python
# ----------------------------------------------
# STEP 6: KMeans Clustering on CAGR and Volatility
# ----------------------------------------------

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Prepare data for clustering
X = sharpe_df[['CAGR (%)', 'Volatility (%)']].values

# Standardize for balanced clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Add cluster labels to DataFrame
sharpe_df['Cluster'] = clusters

# Optional: sort for readability
sharpe_df_sorted_clusters = sharpe_df.sort_values(by='Cluster').reset_index(drop=True)

# Display clustered table
print("✅ KMeans Clustering Completed (4 Groups)")
display(sharpe_df_sorted_clusters)

# ----------------------------------------------
# Visualization of Clusters
# ----------------------------------------------

plt.figure(figsize=(10, 7))

# Scatter plot with cluster coloring
scatter = plt.scatter(
    sharpe_df['CAGR (%)'],
    sharpe_df['Volatility (%)'],
    c=sharpe_df['Cluster'],
    cmap='viridis',
    s=150,
    edgecolor='black',
    alpha=0.8
)

# Annotate with Tickers
for i, ticker in enumerate(sharpe_df['Ticker']):
    plt.annotate(
        ticker,
        (sharpe_df['CAGR (%)'].iloc[i], sharpe_df['Volatility (%)'].iloc[i]),
        textcoords="offset points",
        xytext=(5, 5),
        ha='left',
        fontsize=9
    )

plt.xlabel('CAGR (%)', fontsize=14)
plt.ylabel('Volatility (%)', fontsize=14)
plt.title('AI Stocks Clustering by CAGR and Volatility (KMeans)', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.5)
plt.colorbar(scatter, label='Cluster')
plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------
# STEP 7: Portfolio Weight Suggestion Visualization
# ----------------------------------------------

import matplotlib.pyplot as plt
import numpy as np

# Ensure 'sharpe_df' with 'Cluster' column is available
# If using a different DataFrame name, replace accordingly

# Count the number of stocks in each cluster for proportion
cluster_counts = sharpe_df['Cluster'].value_counts().sort_index()

# Create labels for clarity
cluster_labels = {
    0: 'Cluster 0: Low/Moderate Risk, Good Return',
    1: 'Cluster 1: High Risk, Low/Neg Return',
    2: 'Cluster 2: High Risk, High Return',
    3: 'Cluster 3: Moderate Risk, Modest Return'
}
labels = [cluster_labels[c] for c in cluster_counts.index]

# Suggested weights based on risk profile
weights = {
    0: 0.40,  # Core stable allocation
    1: 0.05,  # Minimal allocation to underperformers
    2: 0.30,  # Aggressive growth
    3: 0.25   # Diversified tilt
}
weight_values = [weights[c] for c in cluster_counts.index]

# ----------------------------------------------
# Pie Chart
# ----------------------------------------------
plt.figure(figsize=(8, 8))
plt.pie(weight_values, labels=labels, autopct='%1.0f%%', startangle=140, textprops={'fontsize': 10})
plt.title('Suggested Portfolio Allocation by Cluster', fontsize=16)
plt.tight_layout()
plt.show()

# ----------------------------------------------
# Bar Chart
# ----------------------------------------------
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, weight_values, color='skyblue', edgecolor='black')
plt.title('Suggested Portfolio Allocation by Cluster', fontsize=16)
plt.ylabel('Portfolio Weight')
plt.xticks(rotation=20, ha='right')

# Annotate weights
for bar, weight in zip(bars, weight_values):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.01,
        f'{weight*100:.0f}%',
        ha='center',
        fontsize=12
    )

plt.tight_layout()
plt.show()
```
